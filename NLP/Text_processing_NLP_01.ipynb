##  Remove punctuation marks:

### 1. using list and isalpha() function

```python
print(len(words_word_tokenize))
print("\n")
no_punc_text=[]

for w in words_word_tokenize:
    if w.isalpha():
        no_punc_text.append(w)
        
print(no_punc_text)
print("\n")
print(len(no_punc_text))
```

## Stopwords removal

### 1. using NLTK

```python
## list of stoopwords
import nltk
from nltk.corpus import stopwords
sw_nltk = stopwords.words('english')
print(sw_nltk)
```

```python
print(len(sw_nltk))
```

```python
## now removing

no_stopword_text=[]
for w in no_punc_text:
    if w not in sw_nltk:
        no_stopword_text.append(w)
        
print(no_stopword_text)
print("\n")

print(len(no_stopword_text))
```

## Stemming 

### 1. porter stemmer

```python
from nltk.stem import PorterStemmer

# Create a Porter Stemmer instance
porter_stemmer = PorterStemmer()


# Apply stemming to each word
stemmed_words = [porter_stemmer.stem(word) for word in no_stopword_text]

# Print the results
print("Original words:", no_stopword_text)
print("\n")
print("Stemmed words:", stemmed_words)
```

### 2. using snowball stemmer


```python
from nltk.stem import SnowballStemmer

# Choose a language for stemming, for example, English
stemmer = SnowballStemmer(language='english')


# Apply Snowball Stemmer
stemmed_words = [stemmer.stem(word) for word in no_stopword_text]

# Print the results
print("Original words:",  no_stopword_text)
print("\n")
print("Stemmed words:", stemmed_words)
```

## lemmatization
### 1. using NLTK's Wordnet Lemmatizer  
Wordnet is a publicly available lexical database of over 200 languages that provides semantic relationships between its words. It is one of the earliest and most commonly used lemmatizer technique.  

It is present in the nltk library in python.
Wordnet links words into semantic relations. ( eg. synonyms )
It groups synonyms in the form of synsets.
synsets : a group of data elements that are semantically equivalent. 

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

# Create WordNetLemmatizer object
wnl = WordNetLemmatizer()


for words in no_stopword_text:
	print(words + " ---> " + wnl.lemmatize(words))
	

```

### 2. Using Spacy

```python
import spacy
nlp = spacy.load('en_core_web_sm')


# Lemmatize each word in the list
lemmatized_words = [token.lemma_ for word in no_stopword_text for token in nlp(word)]

# Print the result
print("Original list of words:", no_stopword_text)
print("Lemmatized words:", lemmatized_words)
```
